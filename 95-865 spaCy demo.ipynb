```python
import spacy
nlp = spacy.load('en_core_web_sm')  # load spaCy's built-in English models
```

```python
text = open('opioid_wikipedia.txt',encoding="utf8").read()  # open text file of text from opioid epidemic Wikipedia page
print(text)
```

```python
parsed_text = nlp(text)
```

```python
for token in parsed_text:
    print(token, token.lemma_, token.pos_, token.ent_type_)
```

```python
for entity in parsed_text.ents:
    print(entity, entity.label_)
```

```python
for idx, sentence in enumerate(parsed_text.sents):
    print('Sentence number', idx, ':', sentence)
```

```python
from collections import Counter
from operator import itemgetter

histogram = Counter()
for token in parsed_text:
    original_token_text = token.orth_  # `token` itself is not a string
    histogram[original_token_text] += 1
#KEY VALUE PAIR WHERE KEY IS WORDS AND VALUE IS COUNTS OF THAT WORD
#lIST OF TUPPLES ARRANGES IN DECREASING ORDER OF COUNTS
sorted_token_count_pairs = sorted(histogram.items(),
                                  reverse=True,
                                  key=itemgetter(1))
for token, count in sorted_token_count_pairs:
    print(token, ":", count)
```

```python
from collections import Counter
from operator import itemgetter

histogram_with_some_filtering = Counter()
for token in parsed_text:
    if not (token.is_stop or token.pos_ == 'PUNCT' or token.pos_ == 'SPACE'):
        lemma = token.lemma_
        histogram_with_some_filtering[lemma] += 1

#dICTIONARY HAS LEMMA AND NOT WORDS NOW
sorted_lemma_count_pairs = sorted(histogram_with_some_filtering.items(),
                                  reverse=True,
                                  key=itemgetter(1))
for lemma, count in sorted_lemma_count_pairs:
    print(lemma, ":", count)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')  # prettier plots

num_top_lemmas_to_plot = 10
top_lemmas = [lemma for lemma, count in sorted_lemma_count_pairs[:num_top_lemmas_to_plot]]
top_counts = [count for lemma, count in sorted_lemma_count_pairs[:num_top_lemmas_to_plot]]
plt.bar(range(num_top_lemmas_to_plot), top_counts, align='center')
plt.xticks(range(num_top_lemmas_to_plot), top_lemmas, rotation=45)
plt.xlabel('Lemma')
plt.ylabel('Raw count')
```
